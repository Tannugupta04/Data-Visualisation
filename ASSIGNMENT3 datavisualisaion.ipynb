```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

#Load Iris Dataset
from sklearn.datasets import load_iris
iris_raw = load_iris()
iris = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
iris['species'] = iris_raw.target_names[iris_raw.target]

iris.head()
```

```python
#BASIC LINE PLOT
plt.figure(figsize=(8, 5), dpi=100)

x = iris.index
y = iris['sepal length (cm)']

plt.plot(x, y, 'g^--', label='Sepal Length')
plt.title('Sepal Length Line Plot', fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Sample Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()
```

```python
#MAR CHART(average petal lenth per species)
plt.figure(figsize=(6, 4), dpi=100)

avg_petal = iris.groupby('species')['petal length (cm)'].mean()
bars = plt.bar(avg_petal.index, avg_petal.values, color='skyblue')

patterns = ['/', 'O', '*']
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

plt.title('Average Petal Length by Species')
plt.ylabel('Length (cm)')
plt.show()
```

```python
#Histogram (Distribution of sepal width)
plt.figure(figsize=(8,5))

plt.hist(iris['sepal width (cm)'], bins=10, color='#abcdef')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Pie Chart (Species Distribution)
plt.figure(figsize=(6,6))

species_counts = iris['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%.1f%%', startangle=140, colors=['#99ff99','#ff9999','#66b3ff'])
plt.title('Distribution of Iris Species')
plt.axis('equal')
plt.show()
```

```python
#Box and Whiskers Plot (All features)

plt.figure(figsize=(10,6))
sns.boxplot(data=iris.drop(columns='species'), palette='Set2')
plt.title('Boxplot of Iris Features')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot  Sepal Length vs Petal Length
plt.figure(figsize=(7,5))
sns.scatterplot(data=iris, x='sepal length (cm)', y='petal length (cm)', hue='species', style='species', s=100)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
#Pair Plot  All Features Colored by Species
sns.pairplot(iris, hue='species', corner=True)
plt.suptitle('Pairwise Plot of Iris Features', y=1.02)
plt.show()
```

```python
# Heatmap  Correlation Matrix
plt.figure(figsize=(8,6))
corr = iris.drop(columns='species').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

```python
# Violin Plot  Distribution + Density
plt.figure(figsize=(8,6))
sns.violinplot(data=iris, y='species', x='petal width (cm)', palette='pastel')
plt.title('Violin Plot of Petal Width by Species')
plt.show()
```

```python
#Swarm Plot  Raw Value Distribution with Species
plt.figure(figsize=(8,6))
sns.swarmplot(data=iris, x='species', y='sepal width (cm)', palette='Set2')
plt.title('Swarm Plot of Sepal Width by Species')
plt.show()
```

```python
#Stacked Bar Chart  Count of Samples by Species (Custom-Made)


plt.figure(figsize=(7,5))
species_counts = iris['species'].value_counts()
plt.bar(species_counts.index, species_counts.values, color=['red','green','gray'])
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# KDE Plot  Kernel Density Estimate for Sepal Length

plt.figure(figsize=(7,5))
for species in iris['species'].unique():
    sns.kdeplot(data=iris[iris['species']==species], x='sepal length (cm)', label=species)
plt.title('KDE Plot of Sepal Length by Species')
plt.legend()
plt.show()
```
